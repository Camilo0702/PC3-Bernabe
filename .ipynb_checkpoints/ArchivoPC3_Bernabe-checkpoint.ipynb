```python
import requests
import zipfile
import io
import pandas as pd

# Definir URLs
dataRepo1 = pd.read_excel('https://github.com/Camilo0702/PC3-Bernabe/raw/refs/heads/main/level1.xlsx')
dataRepo2 = pd.read_excel('https://github.com/Camilo0702/PC3-Bernabe/raw/refs/heads/main/level2.xlsx')
dataRepo3 = pd.read_excel('https://github.com/Camilo0702/PC3-Bernabe/raw/refs/heads/main/level3.xlsx')
dataRepo4 = pd.read_excel('https://github.com/Camilo0702/PC3-Bernabe/raw/refs/heads/main/level4.xlsx')
countryCodesURL = 'https://github.com/Camilo0702/PC3-Bernabe/raw/main/countryCODES.xlsx'
fallecidosURL = 'https://github.com/Camilo0702/PC3-Bernabe/raw/main/fallecidosCOVID.csv'

```

```python
dataRepo1.head()
```

```python
dataRepo1 = dataRepo1[3:].reset_index(drop=True)  # Eliminar filas y reiniciar el índice
# Usar la cuarta fila (ahora la primera) como encabezado
new_header = dataRepo1.iloc[0]  # Tomar la nueva cabecera de la primera fila
dataRepo1 = dataRepo1[1:]  # Eliminar la fila que se convirtió en encabezado
dataRepo1.columns = new_header  # Asignar la nueva cabecera
dataRepo1 = dataRepo1[2:].reset_index(drop=True)  # Eliminar filas y reiniciar el índice
dataRepo1.head()  # Mostrar las primeras filas del DataFrame modificado
```

```python
# Eliminar columnas por índice
dataRepo1 = dataRepo1.drop(dataRepo1.columns[[0, 3, 5, 7, 9, 11, 13, 14]], axis=1)

dataRepo1.head()
```

```python
dataRepo2 = dataRepo2[2:].reset_index(drop=True)  # Eliminar filas y reiniciar el índice
new_header = dataRepo2.iloc[0]  # Tomar la nueva cabecera de la primera fila
dataRepo2 = dataRepo2[0:]  # Eliminar la fila que se convirtió en encabezado
dataRepo2 = dataRepo2.drop(dataRepo2.columns[[0, 6, 8, 10, 11]], axis=1)
dataRepo2.head()  # Mostrar las primeras filas del DataFrame modificado
```

```python
dataRepo3 = dataRepo3[2:].reset_index(drop=True)  # Eliminar filas y reiniciar el índice
new_header = dataRepo3.iloc[0]  # Tomar la nueva cabecera de la primera fila
dataRepo3 = dataRepo3[0:]  # Eliminar la fila que se convirtió en encabezado
dataRepo3 = dataRepo3.drop(dataRepo3.columns[[0, 3, 5, 7, 9, 11, 13, 14]], axis=1)
dataRepo3.head()  # Mostrar las primeras filas del DataFrame modificado
```

```python
new_header = dataRepo4.iloc[0]  # Tomar la nueva cabecera de la primera fila
dataRepo4 = dataRepo4[0:]  # Eliminar la fila que se convirtió en encabezado
dataRepo4 = dataRepo4.drop(dataRepo4.columns[[0, 2, 9]], axis=1)
dataRepo4.head()  # Mostrar las primeras filas del DataFrame modificado
```

```python
dataframes = [dataRepo1, dataRepo2, dataRepo3, dataRepo4]

# Concatenar los DataFrames, ignorando el encabezado duplicado
combined_data = pd.concat(dataframes, ignore_index=True)

# Mostrar las primeras filas del DataFrame combinado
combined_data.head()
```

```python
combined_data
```

```python
combined_data.to_csv("combined_data.csv", index=False)
print("El archivo combined_data.csv ha sido guardado correctamente.")
```

```python
country_codes = pd.read_excel(countryCodesURL)

# Ver las primeras filas del DataFrame de country codes para saber qué columnas tiene
country_codes.head()

```

```python
# Supongamos que quieres hacer el merge usando una columna llamada 'Country' en combined_data
# y 'CountryCode' en country_codes. Puedes cambiar los nombres según sea necesario.

# Realizar el merge
merged_data = pd.merge(combined_data, country_codes, left_on='Country', right_on='Country', how='left')

# Mostrar las primeras filas del DataFrame combinado
merged_data.head()
```

```python
merged_data = merged_data.drop(merged_data.columns[[7, 8, 9, 10]], axis=1)
merged_data.head()
```

```python
merged_data.to_csv("merged_data.csv", index=False)
print("El archivo merged_data.csv ha sido guardado correctamente.")
```

```python
fallecidos_data = pd.read_csv(fallecidosURL, sep=';')
fallecidos_data.head()
```

```python
# Convertir la columna 'FECHA_FALLECIMIENTO' a tipo datetime
fallecidos_data['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_data['FECHA_FALLECIMIENTO'], format='%Y%m%d')

# Agrupar por mes y año y contar el total de fallecidos
# Suponiendo que cada fila representa un fallecido, contaremos las filas por cada mes
reporte_mensual = fallecidos_data.groupby(fallecidos_data['FECHA_FALLECIMIENTO'].dt.to_period('M')).size().reset_index(name='Total_Fallecidos')

# Renombrar la columna del período a un formato más comprensible
reporte_mensual['Mes_Año'] = reporte_mensual['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m')
reporte_mensual = reporte_mensual[['Mes_Año', 'Total_Fallecidos']]  # Reorganizar columnas

# Mostrar el DataFrame resultante
reporte_mensual
```

```python
reporte_mensual.to_csv("reporte_mensual.csv", index=False)
print("El archivo reporte_mensual.csv ha sido guardado correctamente.")
```

```python
fallecidos_data['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_data['FECHA_FALLECIMIENTO'], format='%Y%m%d')

# Agrupar por mes y clasificar por tipo de muerte (suponiendo que el tipo de muerte está en 'CLASIFICACION_DEF')
reporte_mensual_tipo = fallecidos_data.pivot_table(
    index=fallecidos_data['FECHA_FALLECIMIENTO'].dt.to_period('M'),  # Agrupar por mes
    columns='CLASIFICACION_DEF',  # Usar el tipo de muerte como columnas
    aggfunc='size',  # Contar la cantidad de ocurrencias
    fill_value=0  # Rellenar con 0 en caso de que no haya datos para ese tipo
).reset_index()

# Renombrar la columna del período a un formato más comprensible
reporte_mensual_tipo['Mes_Año'] = reporte_mensual_tipo['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m')
reporte_mensual_tipo = reporte_mensual_tipo.drop(columns=['FECHA_FALLECIMIENTO'])  # Eliminar la columna original del periodo

# Mostrar el DataFrame resultante
reporte_mensual_tipo
```

```python
reporte_mensual_tipo.to_csv("reporte_mensual_tipo.csv", index=False)
print("El archivo reporte_mensual_tipo.csv ha sido guardado correctamente.")
```
